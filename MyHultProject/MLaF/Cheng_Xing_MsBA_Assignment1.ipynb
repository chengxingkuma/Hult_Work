```python
# import modules: numpy, pandas, yfinance and datetime
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt
```

```python
# using yf to download multiple tickers into a "data" DataFrame:  
# hint, using yf.download() with these parameters:
# "^GSPC FB AMZN AAPL NFLX GOOG TSLA", start="2000-01-01", end="2021-01-31", interval='1mo'
data = yf.download("^GSPC FB AMZN AAPL NFLX GOOG TSLA", start="2000-01-01", end="2021-01-31", interval='1mo')
data
```

```python
# Drop nan records: these correspond to dividend days, which already accounted for in Adj Close
data=data.dropna()
data
```

```python
#change the date size for easier calculation
year=pd.DataFrame(data.index)['Date'].dt.year
month=pd.DataFrame(data.index)['Date'].dt.month
day=pd.DataFrame(data.index)['Date'].dt.day

#build the yearmon
yearmon=year*100+month
data.index=yearmon

# make an array from Adjusted Close
pr=np.array(data['Adj Close'])

# calculate monthly return and save into a DataFrame df_ret
ret = pr[1:,:]/pr[0:-1,:]-1
df_ret=pd.DataFrame(ret,index=yearmon[1:],columns=data['Adj Close'].columns)
df_ret
```

```python
# Import the linear_model method from sklearn 
from sklearn import linear_model
from sklearn import datasets
```

```python

# Make Training and Validation Data sets, (Xtra, ytra) and (Xval, yval).
# * take first 75% rows for training data, and the rest as validation data.  
# * Your traget (y) should be ^GSPC, and Your "features" (X) should be the FAANG+TSLA stock returns

df = pd.DataFrame(df_ret, columns=["AAPL","AMZN","FB","GOOG","NFLX","TSLA"])
target = pd.DataFrame(df_ret, columns=["^GSPC"])

Xtra=df.iloc[0:76]
Xval=df.iloc[77:102]
ytra=target.iloc[0:76]
yval=target.iloc[77:102]
```

```python
# use Linear_model to train your model (i.e. multiple regression)
lm = linear_model.LinearRegression()
model = lm.fit(Xtra,ytra)
```

```python
# What is the R-square of the model
lm.score(Xtra,ytra) 
```

```python
# Show coefficient
lm.coef_
```

```python
# Show intercept
lm.intercept_
```

```python
# Predict using validation dataset
# Compute the vector of error (residual) of prediction
predictions = lm.predict(Xval)
df_prediction=pd.DataFrame(predictions,index=Xval.index, columns=['Predictions'])
df_prediction['target']=yval['^GSPC']
df_prediction['error']=yval['^GSPC']-predictions[:,0]
df_prediction['squared error']=df_prediction['error']**2
df_prediction
```

```python
# Show out-of-sample R-square, i.e. RSquare of the validation data
lm.score(Xval,yval)
```

```python
# Show the mean of error
np.mean(df_prediction["error"], axis=0)
```

```python
# Show the standard deviation of error
np.std(df_prediction["error"], axis=0)
```

```python
# Show sum of squared error (This is your tracking error)

np.sum(df_prediction['squared error'], axis=0)
```
